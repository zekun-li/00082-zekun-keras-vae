```python
from sklearn import cluster
import numpy as np
valid_Y = np.load('valid_Y.npy')
predict_X = np.load('predict_X.npy')
```

```python
from sklearn import mixture
gmm = mixture.GaussianMixture(n_components=21).fit(predict_X)
```

```python
predict_y = gmm.predict(predict_X)
```

```python
from sklearn.manifold import TSNE as TSNE
X_embedded = TSNE(n_components = 2).fit_transform(predict_X)
```

```python
import matplotlib.pyplot as plt
save_img = 'gmm.jpg'
plt.switch_backend('agg')
plt.figure(figsize=(6, 6)) 
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=valid_Y) 
plt.colorbar() 
plt.savefig(save_img)
from IPython.display import Image 
plt.close()
Image(filename=save_img)
```

```python
save_img = 'gmm_predict_y.jpg'
plt.figure(figsize=(6, 6)) 
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=predict_y) 
plt.colorbar() 
plt.savefig(save_img)
plt.close()
Image(filename=save_img) 
```

```python
import matplotlib.pyplot as plt
from IPython.display import Image 
plt.switch_backend('agg')  
log1 = 'logs/11_lr0.0001_latentdim2048_nomeanfile_sumloss_xscale_train.log'
train_total_err = []
train_l2_err = []
train_kl_err = []
valid_total_err = []
valid_l2_err = []
valid_kl_err = []

ep_idx = []
with open(log1,'r') as f:
    lines = f.readlines()

for line in lines[1:]:
    lsp = line.split(',')
    ep_idx.append(int(lsp[0]))
    train_kl_err.append(float(lsp[1]))
    train_l2_err.append(float(lsp[2]))
    train_total_err.append(float(lsp[3]))
    valid_kl_err.append(float(lsp[4]))
    valid_l2_err.append(float(lsp[5]))
    valid_total_err.append(float(lsp[6]))
    
plt.figure()    
ax = plt.subplot(111)
ax.plot(ep_idx, valid_kl_err, label='valid_kl_loss')
ax.plot(ep_idx, valid_l2_err, label='valid_l2_loss')
ax.plot(ep_idx, valid_total_err, label='valid_total_loss')
plt.title('Validation Loss Curve')
#plt.plot(ep_idx, train_kl_err, 'm--', ep_idx, train_l2_err, 'c--', ep_idx, train_total_err, 'r+',ep_idx, valid_kl_err, 'k--', ep_idx, valid_l2_err, 'y--', ep_idx, valid_total_err, 'g+' )
#plt.plot(ep_idx, valid_kl_err, 'k-', ep_idx, valid_l2_err, 'y-', ep_idx, valid_total_err, 'g-' )
#plt.legend((valid_kl_err, valid_l2_err, valid_total_err), ('valid_kl_err', 'valid_l2_err', 'valid_total_err'))
ax.legend()
plt.savefig('val_loss_curve.jpg')
plt.close()
Image(filename='val_loss_curve.jpg') 
```

```python
plt.figure()
ax1 = plt.subplot(111)
ax1.plot(ep_idx,train_kl_err,label='train_kl_loss')
ax1.plot(ep_idx, valid_kl_err, label='valid_kl_loss')
plt.title('Validation KL Loss Curve')
ax1.legend()
plt.savefig('val_kl_loss_curve.jpg')
plt.close()
Image(filename='val_kl_loss_curve.jpg') 
```
